```python
import numpy
import pandas
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors
from matplotlib import cm
from matplotlib import pyplot
%matplotlib inline
```

```python
icpso_iter_arr = numpy.loadtxt(
    "icpso_history.tsv",
    dtype = "int64",
    delimiter = '\t',
    skiprows = 1,
    usecols = 0
)
```

```python
icpso_score_arr = numpy.loadtxt(
    "icpso_history.tsv",
    dtype = 'float64',
    delimiter = '\t',
    skiprows = 1,
    usecols = 1
)
```

```python
icpso_pos_arr = numpy.loadtxt(
    "icpso_history.tsv",
    dtype = 'float64',
    delimiter = '\t',
    skiprows = 1,
    usecols = [i for i in range(2,2002)]
)
```

```python
hc_sa_set_history = pandas.read_csv(
    "hc_sa_set_history.tsv",
    sep = "\t"
)
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(icpso_pos_arr[:8000,:])
pca.explained_variance_ratio_[0:3].sum()
```

```python
proj_particles = pca.transform(icpso_pos_arr)
```

```python
iter_start = numpy.arange(0,31000,1000)
iter_stop = iter_start + 1000

color_map = colors.LinearSegmentedColormap.from_list("Blue_to_Red",["b","r"])
color_map_norm = colors.Normalize(vmin=0, vmax=30)
color_map_pick = cm.ScalarMappable(norm=color_map_norm, cmap=color_map)
color_map_pick.set_array([])

fig = pyplot.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')
for i,(st,sp) in enumerate(zip(iter_start, iter_stop)):
    ax.scatter(
        proj_particles[st:sp,0],
        proj_particles[st:sp,1], 
        proj_particles[st:sp,2],
        s = 1,
        c = numpy.array([color_map_pick.to_rgba(i)])
        #c=c, marker=m
    )
ax.set_title("ICPSO Swarm movements from 8 iteration decomposition")
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
cbar = fig.colorbar(color_map_pick)
cbar.set_label("Algorithm Iteration")
pyplot.tight_layout()

pyplot.savefig("swarm_iteration.png", dpi=300)
```

```python
color_map = colors.LinearSegmentedColormap.from_list("Blue_to_Red",["r","b"])
icpso_orders = numpy.argsort(-icpso_score_arr)
color_map_norm = colors.Normalize(vmin=icpso_orders.min(), vmax=icpso_orders.max())
color_map_pick = cm.ScalarMappable(norm=color_map_norm, cmap=color_map)
color_map_pick.set_array([])

fig = pyplot.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("ICPSO Swarm movements from 8 iteration decomposition")
ax.scatter(
    proj_particles[:,0],
    proj_particles[:,1], 
    proj_particles[:,2],
    s = 1,
    c = color_map_pick.to_rgba(icpso_orders)
    #c=c, marker=m
)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
cbar = fig.colorbar(color_map_pick)
cbar.set_label("Particle Score Ranking")
pyplot.tight_layout()

pyplot.savefig("swarm_scores.png", dpi=300)
```

```python
# convert hill climber positions to probabilities
def smpl_to_pos(arr, nstates=200):
    l = len(arr)
    pos = numpy.zeros(nstates*l, dtype = 'float64')
    for i,st in enumerate(range(0,l*nstates,nstates)):
        ix = st+arr[i]
        pos[ix] = 1.0
    return pos

groups = hc_sa_set_history.groupby("iter")

hc_best_pos = numpy.zeros((groups.ngroups, 200*10), dtype='float64')
for i,(name,group) in enumerate(groups):
    max_ix = group["score"].argmax()
    tmp = group.iloc[max_ix,2:].values.astype('int64')
    hc_best_pos[i,:] = smpl_to_pos(tmp)
    
hc_pos = numpy.zeros((len(hc_sa_set_history),200*10), dtype='float64')
for i in range(len(hc_sa_set_history)):
    tmp = hc_sa_set_history.iloc[i,2:].values.astype('int64')
    hc_pos[i,:] = smpl_to_pos(tmp)
```

```python
len(hc_sa_set_history)
```

```python
hc_proj_particles = pca.transform(hc_pos)
```

```python
print(hc_proj_particles.shape)
```

```python
pca.fit(numpy.concatenate([icpso_pos_arr, hc_pos]))
print(pca.explained_variance_ratio_[0:3].sum())

icpso_proj_particles = pca.transform(icpso_pos_arr)
hc_proj_particles = pca.transform(hc_best_pos)

iter_start = numpy.arange(0,31000,1000)
iter_stop = iter_start + 1000

color_map = colors.LinearSegmentedColormap.from_list("Blue_to_Red",["b","r"])
color_map_norm = colors.Normalize(vmin=0, vmax=30)
color_map_pick = cm.ScalarMappable(norm=color_map_norm, cmap=color_map)
color_map_pick.set_array([])
particle_colors = color_map_pick.to_rgba(icpso_iter_arr)
hc_colors = color_map_pick.to_rgba([e for e in range(groups.ngroups)])
```

```python
pyplot.rc('font', size=12)          # controls default text sizes
pyplot.rc('axes', titlesize=15)     # fontsize of the axes title
pyplot.rc('axes', labelsize=15)    # fontsize of the x and y labels
pyplot.rc('xtick', labelsize=12)    # fontsize of the tick labels
pyplot.rc('ytick', labelsize=12)    # fontsize of the tick labels
pyplot.rc('legend', fontsize=12)    # legend fontsize
pyplot.rc('figure', titlesize=22)  # fontsize of the figure title
fig = pyplot.figure(figsize=(13,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    icpso_proj_particles[:,0],
    icpso_proj_particles[:,1], 
    icpso_proj_particles[:,2],
    s = 0.1,
    marker = 'o',
    c = particle_colors,
    label = "ICPSO"
    #c=c, marker=m
)
ax.plot(
    hc_proj_particles[:,0],
    hc_proj_particles[:,1],
    hc_proj_particles[:,2],
    "tab:orange",
    label="Hillclimber",
    linewidth = 1
)
ax.scatter(
    hc_proj_particles[:,0],
    hc_proj_particles[:,1],
    hc_proj_particles[:,2],
    s = 6,
    marker = 'o',
    c = hc_colors
)
ax.set_title("ICPSO Swarm Movements vs Steepest Ascent Hillclimber", fontsize=22)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
cbar = fig.colorbar(color_map_pick)
cbar.set_label("Algorithm Iteration", fontsize=15)
ax.legend(loc=1, bbox_to_anchor=(0.8,0.8), fontsize=15)
pyplot.tight_layout()

pyplot.savefig("swarm_iteration_and_hc.png", dpi=300)
```
